```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
gold_data = pd.read_csv('./Downloads/gold_price_data.csv')
print(gold_data.head())
```

```python
gold_data.isnull().sum()
features_X=gold_data.drop(['Date','GLD'],axis=1)
prediction_target_y=gold_data['GLD']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features_X, prediction_target_y, test_size = 0.2, random_state=2)
```

```python
model=RandomForestRegressor(n_estimators=100)
```

```python
model.fit(X_train,Y_train)
```

```python
test_prediction=model.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test,test_prediction )
print("R squared error : ", error_score)
```

```python
Y_test = list(Y_test)
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
plt.plot(Y_test, color='blue', label = 'Actual Value')
plt.plot(test_prediction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
